```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000
```

```python
def _flag(s):
    if s < 10000:
        res = 'A'
    elif s<=100000:
        res = 'B'
    elif s<=200000:
        res = 'C'
    elif s<=400000:
        res = 'D'
    elif s<=700000:
        res = 'E'
    else:
        res = 'F'
    return res
```

```python
df['LB_FLAG'] = df['LIMIT_BAL'].map(lambda s: _flag(s))
#TODO
```

```python
# посмтрим на наши данные в следующих разрезах:
# 1. зависимости SEX, EDUCATION, MERRIAGE и AGE, LIMIT_BAL друг от друга
# 3. зависимости PAY_, BILL_, LIMIT_BAL от SEX, EDUCATION, MERRIAGE и AGE
```

```python
#1) сет данных без PAY и BILL
df_reduced = df.iloc[:,:5]
df_reduced['default.payment.next.month'] = df['default.payment.next.month']
df_reduced.columns
```

```python
plt.figure(figsize=(12, 10))
sns.pairplot(df_reduced, kind = 'scatter', hue='default.payment.next.month')
```

```python
#2) все данные, которые начинаются на PAY, BILL и LIMIT_BAL 
df_p_b = df.iloc[:,5:-2]
df_p_b['LIMIT_BAL'] = df.iloc[:,1] 
df_p_b.columns
```

```python
# строим хитмап
plt.figure(figsize=(12, 10))
sns.heatmap(df_p_b.corr('kendall'), annot=True); #корреляция параметров PAY_ и BILL_ между собой
```

```python
# видно, что PAY_ сильно коррелируют между собой, а также BILL_AMT корредлируют между собой
```

```python
# 3) посмотрим на зависимости PAY_, BILL_, LIMIT_BAL от SEX, EDUCATION, MERRIAGE и AGE
plt.figure()
sns.pairplot(
            df, 
            x_vars = ['SEX','EDUCATION','MARRIAGE','AGE'], 
            y_vars = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1',
                      'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
                      'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5'], 
            kind = 'scatter', 
            hue='default.payment.next.month')
```

```python
# для большей налгядности введем 3 новых переменных (PAY_mean, PAY_AMT_mean и BILL_AMT_mean) и посмотрим на 
# их взаимосвязи с SEX, EDUCATION, MERRIAGE, AGE и LIMIT_BAL

df_reduced['PAY_mean'] = df_p_b.iloc[:,0:6].T.mean()
df_reduced['BILL_AMT_mean'] = df_p_b.iloc[:,6:12].T.mean()
df_reduced['PAY_AMT_mean'] = df_p_b.iloc[:,12:-1].T.mean()
df_reduced.head(2)
```

```python
# строим график
plt.figure(figsize=(12, 10))
sns.pairplot(df_reduced, y_vars = ['SEX','EDUCATION','MARRIAGE','AGE'], x_vars = ['PAY_mean','PAY_AMT_mean','BILL_AMT_mean', 'LIMIT_BAL'], kind = 'scatter', hue='default.payment.next.month')
```
